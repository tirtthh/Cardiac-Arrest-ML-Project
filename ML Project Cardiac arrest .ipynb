```python
data_1 = data.drop('UnderRisk',axis=1)
```

```python
from sklearn.cluster import KMeans
```

```python
kmn = KMeans(n_clusters = 2, )
kmn.fit(data_1)
```

```python
wcss = kmn.inertia_
wcss
```

```python
wcss = []
for i in range(1, 11):
    kmn = KMeans(n_clusters = i, random_state = 1)
    kmn.fit(data_1)
    wcss.append(kmn.inertia_)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x=range(1, 11), y=wcss, marker='o', color='red')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 6, init = 'k-means++', random_state = 2)
y_kmeans = kmeans.fit_predict(data_1)
```

```python
y_kmeans
```

```python
kmeans.cluster_centers_
```

```python
data_1['cluster']=  y_kmeans
data_1.head()
```

```python
data_h = data.drop('UnderRisk',axis=1)
```

```python
from scipy.cluster.hierarchy import linkage,dendrogram
merg = linkage(data_1, method = "ward")

dendrogram(merg, leaf_rotation = 90)
plt.xlabel("data points")
plt.ylabel("euclidean distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters = 6, affinity = "euclidean", linkage = "ward")
cluster = hc.fit_predict(data_h)
```

```python
data_h["label"] = cluster
```

```python
data_h.head()
```

```python
data_h.label.value_counts()
```

```python
from sklearn.metrics import silhouette_score
```

```python
score_agg = silhouette_score(data_h, cluster)
score_agg
```
